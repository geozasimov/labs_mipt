```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
from seaborn import regplot
```

```python
fig, ax = plt.subplots()


plt.title("График зависимости U(I)") # заголовок
plt.xlabel("I") # ось абсцисс
plt.ylabel("U") # ось ординат
ax.grid()      # включение отображение сетки


#  Устанавливаем интервал основных делений:
ax.xaxis.set_major_locator(ticker.MultipleLocator(0.05))
#  Устанавливаем интервал вспомогательных делений:
ax.xaxis.set_minor_locator(ticker.MultipleLocator(0.05))

#  Тоже самое проделываем с делениями на оси "y":
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.1))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.1))


#20 cm
U_up = [0.4395, 0.4991, 0.5556, 0.6955, 0.8105, 1.1800]
I_up = [0.225, 0.255, 0.280, 0.350, 0.405, 0.580]

U_down = [1.1801, 0.8260, 0.7421, 0.5681, 0.5164, 0.4186]
I_down = [0.580, 0.415, 0.365, 0.290, 0.265, 0.215]

U_zero = [0, 0]
I_zero = [0, 0]

ax.scatter(I_up, U_up, color = "red", label = "Точки измерения ВВЕРХ при L = 20 см")
ax.scatter(I_down, U_down, color = "yellow", label = "Точки измерения ВНИЗ при L = 20 см")


U = U_up + U_down
I = I_up + I_down

#zero adding
U = U_zero + U
I = I_zero + I

U = sorted(U)
I = sorted(I)

U_np = np.array(U)
I_np = np.array(I)

m, b = np.polyfit(I_np, U_np, 1)
U_np_right = m * I_np + b

ax.plot(I_np , U_np_right , color = "blue", linewidth = 2, label = "График L = 20 см")

#------------------------------------------------------------------------
#30 cm
U_up = [0.6630, 0.7710, 0.9346, 1.1066, 1.5501, 2.2747]
I_up = [0.225, 0.260, 0.315, 0.370, 0.470, 0.740]

U_down = [1.9452, 1.6787, 1.2231, 1.0467, 0.8342, 0.6921]
I_down = [0.635, 0.505, 0.410, 0.350, 0.280, 0.235]

U_zero = [0, 0]
I_zero = [0, 0]

ax.scatter(I_up, U_up, color = "black", label = "Точки измерения ВВЕРХ при L = 30 см")
ax.scatter(I_down, U_down, color = "cyan", label = "Точки измерения ВНИЗ при L = 30 см")


U = U_up + U_down
I = I_up + I_down

#zero adding
U = U_zero + U
I = I_zero + I

U = sorted(U)
I = sorted(I)

U_np = np.array(U)
I_np = np.array(I)

m, b = np.polyfit(I_np, U_np, 1)
U_np_right = m * I_np + b

ax.plot(I_np, U_np_right, color = "green", linewidth = 2, label = "График L = 30 см")

#------------------------------------------------------------------------
#50 cm
U_up = [1.0666, 1.1646, 1.3515, 1.6372, 2.0074, 3.2820]
I_up = [0.220, 0.235, 0.275, 0.330, 0.400, 0.640]

U_down = [2.7892, 2.3468, 1.8276, 1.5032, 1.2345, 1.1332]
I_down = [0.550, 0.465, 0.365, 0.305, 0.250, 0.230]

U_zero = [0, 0]
I_zero = [0, 0]

ax.scatter(I_up, U_up, color = "purple", label = "Точки измерения ВВЕРХ при L = 50 см")
ax.scatter(I_down, U_down, color = "magenta", label = "Точки измерения ВНИЗ при L = 50 см")


U = U_up + U_down
I = I_up + I_down

#zero adding
U = U_zero + U
I = I_zero + I

U = sorted(U)
I = sorted(I)

U_np = np.array(U)
I_np = np.array(I)

m, b = np.polyfit(I_np, U_np, 1)
U_np_right = m * I_np + b

ax.plot(I_np, U_np_right, color = "orange", linewidth = 2, label = "График L = 50 см")
#------------------------------------------------------------------------

fig.set_figwidth(12)
fig.set_figheight(8)
plt.legend(loc='best')



fname = "schedule.jpg"
plt.savefig(fname, dpi=None, facecolor='w', edgecolor='w',
        orientation='portrait', papertype=None, format=None,
        transparent=False, bbox_inches=None, pad_inches=0.1,
        frameon=None, metadata=None)


plt.show()



```
