```python
fig, ax = plt.subplots()


plt.title("График зависимости v от u") # заголовок
plt.xlabel("u, мc^2") # ось абсцисс
plt.ylabel("v, м^2") # ось ординат
ax.grid()      # включение отображение сетки


#  Устанавливаем интервал основных делений:
ax.xaxis.set_major_locator(ticker.MultipleLocator(0.02))
#  Устанавливаем интервал вспомогательных делений:
ax.xaxis.set_minor_locator(ticker.MultipleLocator(0.02))

#  Тоже самое проделываем с делениями на оси "y":
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.05))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.05))


T = [1.5090,
1.4219,
1.4253,
1.4241,
1.4240,
1.4806,
1.4715,
1.4642,
1.4569,
1.5192]
y = [0.47,
0.18,
0.17,
0.31,
0.31,
0.42,
0.39,
0.59,
0.36,
0.49]

x = [0.350,
0.272,
0.269,
0.306,
0.305,
0.335,
0.329,
0.380,
0.320,
0.355]

#ax.scatter(y, T, color = "red", label = "Точки измерения T(y)")


T_np = np.array(T)
y_np = np.array(y)
x_np = np.array(x)

v_np = y_np ** 2
u_np = T_np ** 2 * x_np


m, b = np.polyfit(u_np, v_np, 1)
v_np_right = m * u_np + b

ax.plot(u_np , v_np_right , color = "blue", linewidth = 2, label = "График зависимости v от u")

#------------------------------------------------------------------------

fig.set_figwidth(12)
fig.set_figheight(8)
plt.legend(loc='best')



fname = "graph_2.jpg"
plt.savefig(fname, dpi=None, facecolor='w', edgecolor='w',
        orientation='portrait', papertype=None, format=None,
        transparent=False, bbox_inches=None, pad_inches=0.1,
        frameon=None, metadata=None)


plt.show()
```
