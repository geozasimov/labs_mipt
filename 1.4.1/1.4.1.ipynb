```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
from seaborn import regplot
from math import sqrt
```

```python
fig, ax = plt.subplots()


plt.title("График зависимости T от y") # заголовок
plt.xlabel("y, м") # ось абсцисс
plt.ylabel("T, с") # ось ординат
ax.grid()      # включение отображение сетки


#  Устанавливаем интервал основных делений:
ax.xaxis.set_major_locator(ticker.MultipleLocator(0.02))
#  Устанавливаем интервал вспомогательных делений:
ax.xaxis.set_minor_locator(ticker.MultipleLocator(0.02))

#  Тоже самое проделываем с делениями на оси "y":
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.01))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.01))


T = [1.5090,
1.4219,
1.4253,
1.4241,
1.4240,
1.4806,
1.4715,
1.4569,
1.5192]
y = [0.47,
0.18,
0.17,
0.31,
0.31,
0.42,
0.39,
0.36,
0.49]

a = [[None] * 2 for i in range(len(T))]

for i in range(len(T)):
    a[i][0] = y[i]
    a[i][1] = T[i]


a = sorted(a, key=lambda a: a[0])
print(a)

T_new = [None] * len(T)
y_new = [None] * len(y)

for i in range(len(T)):
    y_new[i] = a[i][0]
    T_new[i] = a[i][1]

ax.scatter(y, T, color = "red", label = "Точки измерения T(y)")
print(T_new)
print(y_new)

T_np = np.array(T_new)
y_np = np.array(y_new)

ax.plot(y_np , T_np , color = "blue", linewidth = 2, label = "График завимости периода колебаний Т от смещения груза y")

#m, b = np.polyfit(y_np, T_np, 1)
#T_np_right = m * y_np + b

#ax.plot(y_np , T_np_right , color = "blue", linewidth = 2, label = "График завимости периода колебаний Т от смещения груза y")

#------------------------------------------------------------------------

fig.set_figwidth(12)
fig.set_figheight(8)
plt.legend(loc='best')



fname = "graph_1.jpg"
plt.savefig(fname, dpi=None, facecolor='w', edgecolor='w',
        orientation='portrait', papertype=None, format=None,
        transparent=False, bbox_inches=None, pad_inches=0.1,
        frameon=None, metadata=None)


plt.show()



```

```python
fig, ax = plt.subplots()


plt.title("График зависимости v от u") # заголовок
plt.xlabel("u, мc^2") # ось абсцисс
plt.ylabel("v, м^2") # ось ординат
ax.grid()      # включение отображение сетки


#  Устанавливаем интервал основных делений:
ax.xaxis.set_major_locator(ticker.MultipleLocator(0.02))
#  Устанавливаем интервал вспомогательных делений:
ax.xaxis.set_minor_locator(ticker.MultipleLocator(0.02))

#  Тоже самое проделываем с делениями на оси "y":
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.05))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.05))


T = [1.5090,
1.4219,
1.4253,
1.4241,
1.4240,
1.4806,
1.4715,
1.4642,
1.4569,
1.5192]
y = [0.47,
0.18,
0.17,
0.31,
0.31,
0.42,
0.39,
0.59,
0.36,
0.49]

x = [0.350,
0.272,
0.269,
0.306,
0.305,
0.335,
0.329,
0.380,
0.320,
0.355]




T_np = np.array(T)
y_np = np.array(y)
x_np = np.array(x)

v_np = y_np ** 2
u_np = T_np ** 2 * x_np
ax.scatter(u_np, v_np, color = "red", label = "Точки измерения v от u")

m, b = np.polyfit(u_np, v_np, 1)
v_np_right = m * u_np + b

#ax.plot(u_np , v_np , color = "green", linewidth = 2, label = "График зависимости v от u")
ax.plot(u_np , v_np_right , color = "blue", linewidth = 2, label = "График зависимости v от u")

#------------------------------------------------------------------------

fig.set_figwidth(12)
fig.set_figheight(8)
plt.legend(loc='best')



fname = "graph_2.jpg"
plt.savefig(fname, dpi=None, facecolor='w', edgecolor='w',
        orientation='portrait', papertype=None, format=None,
        transparent=False, bbox_inches=None, pad_inches=0.1,
        frameon=None, metadata=None)


plt.show()
```
