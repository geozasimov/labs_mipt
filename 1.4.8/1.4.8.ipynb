```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
from seaborn import regplot
from math import sqrt
```

```python
fig, ax = plt.subplots()


plt.title("График зависимости резонансной частоты стержня f_n от номера гармоники n") # заголовок
plt.xlabel("n, А") # ось абсцисс
plt.ylabel("f_n, В") # ось ординат
ax.grid()      # включение отображение сетки


#  Устанавливаем интервал основных делений:
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
#  Устанавливаем интервал вспомогательных делений:
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))

#  Тоже самое проделываем с делениями на оси "y":
ax.yaxis.set_major_locator(ticker.MultipleLocator(2000))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(2000))


#al
f_n = [0,
4262.25,
8526.93,
12790.9,
17053.1,
21303.4,
25574.4]

n = [i for i in range(len(f_n))]


ax.scatter(n, f_n, color = "red", label = "Точки измерения для алюминия")

f_n_np = np.array(f_n)
n_np = np.array(n)

m, b = np.polyfit(n_np, f_n_np, 1)
f_n_np_right = m * n_np + b

ax.plot(n_np , f_n_np_right , color = "red", linewidth = 2, label = "График зависимости f_n от n для алюминия")

#------------------------------------------------------------------------

#med
f_n = [0,
3218.89,
6422.54,
9646.03,
12854.4,
16064.6,
18255.6]

n = [i for i in range(len(f_n))]


ax.scatter(n, f_n, color = "green", label = "Точки измерения для меди")

f_n_np = np.array(f_n)
n_np = np.array(n)

m, b = np.polyfit(n_np, f_n_np, 1)
f_n_np_right = m * n_np + b

ax.plot(n_np , f_n_np_right , color = "green", linewidth = 2, label = "График зависимости f_n от n для меди")

#------------------------------------------------------------------------

#stal
f_n = [0,
4132.75,
8257.89,
12392.7,
16522  ,
20645.3,
24758.9]

n = [i for i in range(len(f_n))]


ax.scatter(n, f_n, color = "blue", label = "Точки измерения для стали")

f_n_np = np.array(f_n)
n_np = np.array(n)

m, b = np.polyfit(n_np, f_n_np, 1)
f_n_np_right = m * n_np + b

ax.plot(n_np , f_n_np_right , color = "blue", linewidth = 2, label = "График зависимости f_n от n для стали")

#------------------------------------------------------------------------

fig.set_figwidth(12)
fig.set_figheight(8)
plt.legend(loc='best')



fname = "schedule.jpg"
plt.savefig(fname, dpi=None, facecolor='w', edgecolor='w',
        orientation='portrait', papertype=None, format=None,
        transparent=False, bbox_inches=None, pad_inches=0.1,
        frameon=None, metadata=None)


plt.show()



```
