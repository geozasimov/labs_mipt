```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
from seaborn import regplot
from math import sqrt
```

```python
fig, ax = plt.subplots()


plt.title("График зависимости скорости прецессии от момента силы") # заголовок
plt.xlabel("M, нм") # ось абсцисс
plt.ylabel("Omega, с^-1") # ось ординат
ax.grid()      # включение отображение сетки


#  Устанавливаем интервал основных делений:
ax.xaxis.set_major_locator(ticker.MultipleLocator(0.05))
#  Устанавливаем интервал вспомогательных делений:
ax.xaxis.set_minor_locator(ticker.MultipleLocator(0.05))

#  Тоже самое проделываем с делениями на оси "y":
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.025))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.025))


#20 cm
Om = [
0.05651,     
0.07145,     
0.08874, 
0.10698, 
0.12604, 
0.16580]

M = [
0.1094796,   
0.1365552,
0.1671624,
0.2036556,
0.2519208,
0.317844]

ax.scatter(M, Om, color = "red", label = "Точки измерения зависимости скорости прецессии от момента силы")

M = [0] + M
Om = [0] + Om
Om_np = np.array(Om)
M_np = np.array(M)

m, b = np.polyfit(M_np, Om_np, 1)
Om_np_right = m * M_np + b

ax.plot(M_np , Om_np_right , color = "blue", linewidth = 2, label = "График зависимости скорости прецессии от момента силы")

#------------------------------------------------------------------------

fig.set_figwidth(12)
fig.set_figheight(8)
plt.legend(loc='best')



fname = "OmegaM.jpg"
plt.savefig(fname, dpi=None, facecolor='w', edgecolor='w',
        orientation='portrait', papertype=None, format=None,
        transparent=False, bbox_inches=None, pad_inches=0.1,
        frameon=None, metadata=None)


plt.show()



```
